## K-Means Clustering Algorithm
K-Means Clustering is an unsupervised learning algorithm that is used to solve the clustering problems in machine learning or data science. In this topic, we will learn what is K-means clustering algorithm, how the algorithm works, along with the Python implementation of k-means clustering.


### Clustering Vs Classification :
Clustering is often confused with classification, which is another technique used in machine learning. However, there are some key differences between the two.

Classification is a supervised learning method that involves assigning predefined classes or labels to data points based on their features or attributes. In contrast, clustering is an unsupervised learning method that groups data points based on their similarities.

While classification is used to predict the class or label of new data points, clustering is used to identify patterns and relationships in data without any predefined classes or labels.

#####  K-Means Clustering is an Unsupervised Learning algorithms, which groups the Unlabeled data into different Clusters.
#####  K defines the number of pre-defined clusters that need to be created in the process.
#####  K-means clustering , cluster the data on the basis of MEAN
#####  First K-mean Clustering decides it's Centeroid, it is a centre point.
#####  After applying mean to make another new centre point Then joins the two centre lines,and passes another line from it's centre to divide the cluster perfectly.
#####  The centeroid point find it's nearest point and make a Cluster , same as another centre point will also find it's nearest point to make Cluster.

### How does the K-Means Algorithm Work?

#### Step-1: Select the number K to decide the number of clusters.

#### Step-2: Select random K points or centroids. (It can be other from the input dataset).

#### Step-3: Assign each data point to their closest centroid, which will form the predefined K clusters.

#### Step-4: Calculate the variance and place a new centroid of each cluster.

#### Step-5: Repeat the third steps, which means reassign each datapoint to the new closest centroid of each cluster.

#### Step-6: If any reassignment occurs, then go to step-4 else go to FINISH.

#### Step-7: The model is ready.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_csv('iris.csv')
dataset.head(5)
```

```python
# This dataset has no output , k-means cluster will tell us how many cluster will make in this dataset
```

```python
sns.pairplot(data=dataset)
plt.show()
```

##### here we can observe that it shows only two cluster of data , but we know that the original dataset contain 3 cluster of output , but we cannot see it in this graph so, K=MEAN CLUSTER will help to find 3 no. of clusters.

```python
from sklearn.cluster import KMeans
```

```python
wcss = []
# Loop through a range of cluster numbers
for i in range(2, 21):
    km = KMeans(n_clusters=i, init='k-means++', random_state=42)  # Random state for reproducibility
    km.fit(dataset)  # Fit KMeans to the dataset
    wcss.append(km.inertia_)  # Append the WCSS value for each cluster
```

```python
# Plotting the results to observe the "Elbow" for optimal clusters
plt.plot([i for i in range(2, 21)], wcss, marker='o')
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.xticks([i for i in range(2, 21)])
plt.grid(axis='x')
plt.ylabel('WCSS')
plt.show()
```

```python
# Elbow point here is 3, so it means there is 3 no. of clusters is going to make.
```

```python
km = KMeans(n_clusters=3)
dataset['predict']=km.fit_predict(dataset)
```
