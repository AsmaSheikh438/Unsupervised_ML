### DBSCAN CLUSTERING ALGORITHM : 
#### Density-Based Spatial Clustering of Application with noise. The clusters found by DBSCAN can be any shape , can deal with some special cases that other methods cannot

#### DBSCAN CLUSTERING used for non linear seprable data

```python
import matplotlib.pyplot as plt
import seaborn as sns 
from sklearn.datasets import make_moons
import pandas as pd 
```

```python
x,y = make_moons(n_samples=250, noise=0.05)
```

```python
x
```

```python
y
```

```python
df = {'data1': x[:, 0], 'data2': x[:, 1], 'output': y}
```

```python
x[:,0]
```

```python
dataset = pd.DataFrame(df)
```

```python
dataset.head(3)
```

```python
# chceking the graph is linear or not by plotting graph
```

```python
sns.scatterplot(x='data1', y='data2', data=dataset)
plt.show()
```

```python
from sklearn.cluster import DBSCAN
db = DBSCAN(eps=0.2, min_samples=5)
dataset['predict'] = db.fit_predict(dataset)
```

```python
dataset.head(3)
```

```python
sns.scatterplot(x='data1', y='data2', data=dataset, hue='predict')
plt.show()
```
